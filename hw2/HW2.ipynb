### Problem 1

$s(\theta, \phi) = (\cos\theta\cos\phi, \sin\theta\cos\phi, \sin\phi)$

#### a) 
Equations for coordinate tangent vectors $\frac{\partial s}{\partial \theta}, \frac{\partial s}{\partial \phi}$:

$ \frac{\partial s}{\partial \theta} = (-\sin\theta\cos\phi, \cos\theta\cos\phi, 0) $ 
$\frac{\partial s}{\partial \phi} = (-\cos\theta\sin\phi, -\sin\theta\sin\phi, \cos\phi)$


#### b) 
Coefficients of Riemannian metric, $g$, on $S^2$:

$g_{ij} = \langle v^i, v^j \rangle$, where $v^i = \frac{\partial}{\partial x^i}$ and $x^1 = \theta, x^2 = \phi$

$\begin{align*}
&g = \begin{matrix}
g_{11} & g_{12}\\
g_{21} & g_{22}
\end{matrix} \\
&= \begin{matrix}
\langle \frac{\partial s}{\partial \theta}, \frac{\partial s}{\partial \theta}\rangle & \langle \frac{\partial s}{\partial \theta}, \frac{\partial s}{\partial \phi}\rangle\\
\langle \frac{\partial s}{\partial \phi}, \frac{\partial s}{\partial \theta}\rangle & \langle \frac{\partial s}{\partial \phi}, \frac{\partial s}{\partial \phi}\rangle
\end{matrix} \\
&= \begin{matrix}
\cos^2\phi & 0\\
0 & 1
\end{matrix}
\end{align*}$

#### c)
Inverse metric $g^{-1}$:

$\begin{align*}
&g^{-1} = \frac{1}{g_{11}g_{22} - g_{12}g_{21}} \begin{matrix}
g_{22} & -g_{12}\\
-g_{21} & g_{11}
\end{matrix} \\
&g^{-1} = \begin{matrix}
\frac{1}{\cos^2\phi} & 0\\
0 & 1
\end{matrix}
\end{align*}$

#### d)
Formulas for Christoffel symbols, $\Gamma_{ij}^k$

$\Gamma_{ij}^k = \frac{1}{2} \sum_{\ell=1}^{n} g^{k\ell} \left( \frac{\partial g_{j\ell}}{\partial x^i} + \frac{\partial g_{i\ell}}{\partial x^j} - \frac{\partial g_{ij}}{\partial x^\ell} \right)$


$\Gamma_{\theta\theta}^\theta = \frac{1}{2}\left[\frac{1}{\cos^2\theta}\left(0 - 2\sin\phi\cos\phi - 0\right) + 0\right] = \frac{-2\sin\phi\cos\phi}{2\cos\phi} = -\tan\phi$

$\Gamma_{\theta\theta}^\phi = \frac{1}{2}\left[0 + 1\left(0 + 0 + 2\sin\phi\cos\phi\right)\right] = \sin\phi\cos\phi$

$\Gamma_{\theta\phi}^\theta = \frac{1}{2}\left[\frac{1}{\cos^2\phi}\left(0 -2\sin\phi\cos\phi - 0\right) + 0\right] = -\tan\phi$

$\Gamma_{\theta\phi}^\phi = \frac{1}{2}\left[0 + 1(0 + 0 - 0)\right] = 0$

$\Gamma_{\phi\theta}^\theta = \frac{1}{2}(0 + 0) = 0$

$\Gamma_{\phi\theta}^\phi = \frac{1}{2}\left[0 + 1(0 + 0 - 0)\right] = 0$

$\Gamma_{\phi\phi}^\theta = \frac{1}{2}\left[0 + 1(0 + 0 - 0)\right] = 0$

$\Gamma_{\phi\phi}^\phi = \frac{1}{2}\left[0 + 1(0 + 0 - 0)\right] = 0$


#### e)
Show that the equator circle $\gamma(t) = s(t,0)$ is a geodesic

$\gamma^\theta(t) = t, \gamma^\phi(t) = 0 \implies \frac{d\gamma^\theta}{dt} = 1 , \frac{d\gamma^\phi}{dt} = 0$

$\begin{align*}
&\frac{d^2\gamma^k}{dt^2} = -\sum_{i,j}^{n}\Gamma_{i,j}^{k}\left(\gamma\left( t \right) \right)\frac{d\gamma^i}{dt}\frac{d\gamma^j}{dt}\\
\end{align*}$

$\begin{align*}
&\frac{d^2\gamma^\theta}{dt^2} = -\left( \Gamma_{\theta\theta}^{\theta}\frac{d\gamma^\theta}{dt}\frac{d\gamma^\theta}{dt} + \Gamma_{\theta\phi}^{\theta}\frac{d\gamma^\theta}{dt}\frac{d\gamma^\phi}{dt} +\Gamma_{\phi\theta}^{\theta}\frac{d\gamma^\phi}{dt}\frac{d\gamma^\theta}{dt} +\Gamma_{\phi\phi}^{\theta}\frac{d\gamma^\phi}{dt}\frac{d\gamma^\phi}{dt} \right)\\
&= \tan\phi \\
& \tan0 = 0 \\
& \frac{d^2\gamma^\phi}{dt^2} = 0\\
\end{align*}$

$\begin{align*}
&\frac{d^2\gamma^\phi}{dt^2} = -\left( \Gamma_{\theta\theta}^{\phi}\frac{d\gamma^\theta}{dt}\frac{d\gamma^\theta}{dt} + \Gamma_{\theta\phi}^{\phi}\frac{d\gamma^\theta}{dt}\frac{d\gamma^\phi}{dt} +\Gamma_{\phi\theta}^{\phi}\frac{d\gamma^\phi}{dt}\frac{d\gamma^\theta}{dt} +\Gamma_{\phi\phi}^{\phi}\frac{d\gamma^\phi}{dt}\frac{d\gamma^\phi}{dt} \right)\\
&= -\sin\phi\cos\phi\\
& -\sin0\cos0 = 0 \\
& \frac{d^2\gamma^\phi}{dt^2} = 0\\
\end{align*}$

Since $\frac{d^2\gamma^\phi}{dt^2}, \frac{d^2\gamma^\theta}{dt^2} = 0$, we know that the equator circle minimizes the energy function and is therefore a geodesic. 

#### f)
Every great circle on $S^2$ can be thought of as some rotation of the equator. Since a simple rotation preserves the Riemannian metric, we know that every great circle is an isometry of the equator. Since the equator is a geodesic (i.e it minimizes the Energy function) and the image of a geodesic under an isometry is also geodesic, every great circle must again be a geodesic.

##### Orthogonal Procrustes Analysis

```python
def opa(a, b):
    # a = target, b = moving
    c = np.matmul(b,a.T)
    u, s, vt = np.linalg.svd(c, full_matrices=False)
    r = np.matmul(u,vt) 
    return r
```

##### Exponential and Log Map

```python
import math

'''
Exponential map: given a preshape and tangent vector, 
compute the preshape at the end of the corresponding geodesic
'''
def exp_map(a,v):
    theta = np.linalg.norm(v) # theta = ||v||
    if theta == 0:
        return a
    b = math.cos(theta)*a + math.sin(theta)*(v/theta)
    return b
    
'''
Log map: given two preshapes, compute the tangent vector at the first preshape 
that is the initial condition of the geodesic segment between the two
'''
def log_map(a,b):
    # arc length along unit sphere
    if (np.linalg.norm(a)*np.linalg.norm(b) == 0):
        return np.zeros((2,3))
    x = np.sum(a*b)/(np.linalg.norm(a)*np.linalg.norm(b))
    if x > 1:
        x = 1
    if x < -1:
        x = -1
    theta = math.acos(x)
    # unit normalization 
    v = (b - math.cos(theta)*a)/(np.linalg.norm(b - math.cos(theta)*a))
    # scaling by velocity
    v = v * theta
    return v
```

```python
# Estimating Frechet mean with gradient descent
nsteps = 500

def frechet_mean(shape_arr, nsteps=500):
    u = shape_arr[0]
    for i in range(nsteps):
        s = 0
        for shape in shape_arr:
            r = opa(u, shape)
            ra = np.matmul(r,shape)
            s += log_map(u, ra)
        dv = s/len(shape_arr)
        u = exp_map(u, dv)
        u = u/np.linalg.norm(u)
    return u

u = frechet_mean(triangle_arr)
```

##### Principal Geodesic Analysis (Tangent Approximation) for Triangle Dataset

```python
# Principal Geodesic Analysis

m = frechet_mean(triangle_arr)

mapped_pts = []

for i in range(len(triangle_arr)):
    r = opa(m, triangle_arr[i])
    ra = np.matmul(r,triangle_arr[i])
    u = log_map(m, ra)
    u = np.reshape(u, (1,6))
    prod = np.matmul(np.transpose(u), u)
    mapped_pts.append(prod)

mapped_pts = np.array(mapped_pts)
    
mapped_pts = (1/(len(triangle_arr)-1))*np.sum(mapped_pts, axis=0) 

u, s, vt = np.linalg.svd(np.transpose(mapped_pts))

print(u.shape)
print(s.shape)
```

#### Summary and Analysis

In order to approximate PGA, we are projecting all of our shapes into the tangent space of the Frechet mean. We are doing this by computing $\log_{\bar{y}}y_i$ for every shape $y_i$ and effectively performing PCA in the tangent space. Our principal components will be the tangent vectors that capture the largest variance when the data is projected onto them. 

Because of the loss in degrees of freedom imposed by our pre-processing steps (centering our data around the origin and normalizing) we are ensuring that our data lives on the unit sphere. 

In order to compute the Frechet mean, which is essential for finding the covariance matrix and performing PCA, I had to write three primary functions: OPA and the exponential and log maps. These are important for implementing gradient descent as they allow me to compute geodesics along the unit sphere between the current guess for the intrinsic mean and every shape in the dataset at each timestep. Since our manifold is essentially $S^2$, finding the exponential and log maps turned into a much simpler trigonometry problem, as shown by the diagram below. 

<img src="./diagram.png" width="400" height="400" align="left"/>


Once I found the Frechet mean, I was able to approximate PGA by subtracting the mean from each shape (aligning each shape with the mean and finding the geodesic between them) and creating a covariance matrix. I used SVD to decompose this positive semi-definite matrix and obtain the eigenvectors and eigenvalues. The eigenvalues represent the magnitude of the variance along each of the corresponding eigenvectors.

I generated scree plots for the eigenvalues and also made plots of accumulated data variance vs. number of principal components. In order to do this, I set a threshold of 90% (the percent of variation that I want to capture with my selected components). Numpy's SVD function returns the eigenvalues in descending order so I summed all of the eigenvalues and kept track of a ratio where the denominator is the total sum and the numerator is the partial sum of the first $k$ eigenvalues. I started with the largest eigenvalue and repeatedly added to it, checking if that ratio exceeded 0.9. 

For the triangle dataset, I found that we can capture 90% of the variance with only 2 principal components, and for the corpus collosum dataset, I found that we can capture 90% of the variance with just 6 principal components. 

If we move along the geodesic from the Frechet mean specified by the tangent vector (eigenvector), we are obtaining maximal variation in that direction. In order to find and plot the different modes of variation, I multiplied the eigenvectors we were interested in by the standard deviation (square root of the corresponding eigenvalue) and scaled that up or down in the range $\alpha \in [-2,2]$. Since the eigenvector is a velocity, when we scale it by $\alpha$, we can use the exponential map to traverse our principal components and see how the mean shape is deformed at various standard deviations along each component. 

For the triangle dataset, the plots of the largest modes of variation appear to correspond directly to the steps we took to initialize our dataset (introducing Gaussian noise). The first mode of variation appears to capture variance in the y-direction, which makes sense. Recall how we are creating our dataset:

We are initializing three points for each triangle: $p_0 = (-1,0), p_1 = (1,0), p_2 = (0,s)$.  

While we are introducing some very slight Gaussian noise to all six coordinates of every triangle ($\sigma = 0.1$), there is not much perturbation between the points along the base between any two triangles. However, $s$ is a Gaussian random variable with $\sigma = 0.5$, meaning that the majority of the variation between any two shapes is in the y-direction.

The second mode of variation for the triangle dataset appears to capture the variance introduced by the noise that was applied to the base points ($p_0$ and $p_1$) and the third variation appears to not provide much new information -- it seems to be capturing some combination of the variance described by the first two modes. This makes sense because when calculating the cumulative variance, I found that we only need 2 principal componenents to capture 90% of the variance for the triangle dataset, making the third mode extraneous for the most part. 

For the corpus collosum dataset, it is a bit harder to identify direct correlations between the modes of variation and the raw shape data but we can definitely see that each mode is capturing variance in different areas of the corpus collosum (e.g on the left side, towards the top, or on the right side).

In general, I would expect the maximal number of modes of variance for a given dataset to simply be $d\cdot k$, where $k$ is the number of points in each shape and $d$ is the dimension of our shape (e.g each shape lies in $\mathbb{R}^d$). If we have fewer shapes in our dataset than there are points in each shape, the upper bound on the number of principal components we will need will simply be the number of shapes in the dataset, since each can be adequately represented by a single principal component.
